```python
# Без использования numpy
def mult(a, b):
    result = [[0 for _ in range(len(b[0]))] for _ in range(len(a))]
    for i in range(len(a)):
        for j in range(len(b[0])):
            for k in range(len(b)):
                result[i][j] += a[i][k] * b[k][j]
    return result
```

```python
# С использованием numpy
def np_mult(a, b):
    result = np.dot(a, b)
    return result
```

```python
%%time
# засечем время работы функции без NumPy
M1 = mult(a, b)
```

```python
%%time
# засечем время работы функции с NumPy
M2 = np_mult(a, b)
```

```python
# проверим корректность
np.abs(np.array(M1) - M2).sum()
```

```python
# Скалярное произведение без NumPy
def scalar_product(a, b):
    result = 0
    for i in range(len(a)):
        result += a[i] * b[i]
    return result
```

```python
# Скалярное произведение с использованием NumPy
def np_scalar_product(a, b):
    return np.dot(a, b)
```

```python
%time product_1 = scalar_product(a,b)
%time product_2 = np_scalar_product(a,b)
# проверим корректность:
np.abs(product_1 - product_2).sum()
```

```python
#ЗАДАНИЕ 3

def np_diag_2k(matrix):
    diagonal_elements = np.diag(matrix, k=0)
    even_diagonal_elements = diagonal_elements[::2]

    if even_diagonal_elements.size > 0:
        return np.sum(even_diagonal_elements)
    else:
        return 0
```

```python
#ЗАДАНИЕ 4

def np_sec_av(*A):
    X = np.vstack(A)
    cumsum_matrix = np.cumsum(X, axis=1)
    row_indices = np.arange(1, X.shape[1] + 1)
    result = cumsum_matrix / row_indices[:, np.newaxis]
    return result
```

```python
# зададим некоторую последовательность и проверим ее на ваших функциях.
# функция должна работать ~ в 50 раз быстрее, чем без Numpy
A = sps.uniform.rvs(size=10**3)
print(A)

%time S2 = np_sec_av(A)
```

```python
#ЗАДАНИЕ 5
def np_encode(a):
    grouped_data = [(key, len(list(group))) for key, group in groupby(a)]
    x, num = zip(*grouped_data)
    x = np.array(x)
    num = np.array(num)
    return x, num
```

```python
from itertools import groupby
X = np.array([1, 2, 2, 3, 3, 1, 1, 5, 5, 2, 3, 3])
%time x, num = np_encode(X)
```

```python
#ЗАДАНИЕ 6
def np_transformation(X, a=1):
    result = X.copy()
    result[1::2] = a
    result[0::2] = result[0::2] ** 3
    result = result[::-1]
    final_result = np.concatenate((X, result))

    return final_result
```

```python
X = np.array([i for i in range(1, 10, 2)])
# функция должна работать ~ в 20 раз быстрее, чем без Numpy
%time S2 = np_transformation(X, 5)
```
